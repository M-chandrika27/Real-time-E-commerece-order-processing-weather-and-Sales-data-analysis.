**1.Real-time E-commerece-order-processing.**

##sample data

```python
product1_name='Laptop'
product1_price=1000
product1_stock=5
product2_name='Phone'
product2_price=500
product2_stock=10
product3_name='HeadPhones'
product3_price=100
product3_stock=50
order1_product='Laptop'
order1_quantity=1
order2_product='Phone'
order2_quantity=2
order3_product='HeadPhones'
order3_quantity=3
customer_name='John Doe'
customer_loyalty_points=150
customer_is_member=True
```

#constants and initalize variables

```python
#constants
DISCOUNT_RATE=0.1
POINTS_RATE=0.01
MINIMUM_POINTS_FOR_DISCOUNT=100
#initialize variables
total_price=0
```

#order processing

```python
#process order1
if order1_product==product1_name and product1_stock>=order1_quantity:
  order1_price=product1_price*order1_quantity
  if customer_is_member:
    order1_price-=order1_price*DISCOUNT_RATE
  total_price+=order1_price
  product1_stock-=order1_quantity
else:
  print(f"sorry,we can't fulfull the order for{order1_product}.Not enough stock or product does not exist.")
#process order2
if order2_product==product2_name and product2_stock>=order2_quantity:
  order2_price=product2_price*order2_quantity
  if customer_is_member:
    order2_price-=order2_price*DISCOUNT_RATE
  total_price+=order2_price
  product2_stock-=order2_quantity
else:
  print(f"sorry,we can't fulfull the order for{order2_product}.Not enough stock or product does not exist.")
#process order3
if order3_product==product3_name and product3_stock>=order3_quantity:
  order3_price=product3_price*order3_quantity
  if customer_is_member:
    order3_price-=order3_price*DISCOUNT_RATE
  total_price+=order3_price
  product3_stock-=order3_quantity
else:
  print(f"sorry,we can't fulfull the order for{order3_product}.Not enough stock or product does not exist.")
```

#calculating loyalty points earned

```python
#calculating loyalty points earned
Loyalty_points_earned=int(total_price*POINTS_RATE)
customer_loyalty_points+=Loyalty_points_earned
```

#Apply additional discount if customer has enough loyality points

```python
if customer_loyalty_points>=MINIMUM_POINTS_FOR_DISCOUNT:
  total_price-=total_price*DISCOUNT_RATE
print("Additional discount applied due to loyalty points.")
```

#summary

```python
#summary
print(f"order processed for:{customer_name}")
print(f"Total price:${total_price:.2f}")
print(f"loyalty points earned:{Loyalty_points_earned}")
print(f"Total loyalty points:{customer_loyalty_points}")
print(f"updated product stock:{product1_name}={product1_stock},{product2_name}={product2_stock},{product3_name}={product3_stock}")
```

summary:

This Python-driven real-time e-commerce system processes orders instantly, as seen with John Doe's $1863.00 purchase. It simultaneously updates stock (e.g., Laptop to 4, Phone to 8), calculates new loyalty points, and applies discounts, streamlining fulfillment and inventory management.

**2.Weather data Analysis**

**Real-Life Implementation Scenario :-** weather data analysis.

**Scenario description:-**we are going to analyze weather data using numpy.we will perform various operations such as loading data, basic analysis, statistical operations and reshaping data.

#import libraries

```python
import numpy as np
```

#sample weather data:Temperature readings(i=Celsius)for a week

```python
import numpy as np
weather_data=np.array([[23,25,22,20,24,26,27],#monday
    [22,24,23,19,22,25,26],#tuesday
    [21,23,20,18,23,24,25],#wednessday
    [20,22,19,17,21,23,24],#thrusday
    [19,21,18,16,20,22,23],#friday
    [18,20,17,15,19,21,22],#satruday
    [17,19,16,14,18,20,21]])#sunday
print("weather data:\n",weather_data)
```

step-2:Basic analysis

```python
#average temperature for each day
average_temp_daily=np.mean(weather_data,axis=1)
print("average temperature for each day:\n",average_temp_daily)
```

```python
#average temperature for each week
average_temp_weekly=np.mean(weather_data)
print("average temperature for each week:\n",average_temp_weekly)
```

step-3:statistical opreation

```python
#maximum temperature for each day
max_temp_daily=np.max(weather_data,axis=1)
print("maximum temperature for each day:\n",max_temp_daily)
```

```python
#manimum temperature for each day
min_temp_daily=np.min(weather_data,axis=1)
print("minimum temperature for each day:\n",min_temp_daily)
```

```python
#standard derivation of temperature for the week
std_temp_weekly=np.std(weather_data)
print("standard deviation of weekly temperature:",std_temp_weekly)
```

step-4:reshaping data

```python
#flatten the data to perform element wise operations
weather_data_flattened=weather_data.flatten()
print("flattened weather data:\n",weather_data_flattened)
```

```python
#reshaping the flattened data back to its orginal shape
weather_data_reshaped=weather_data_flattened.reshape(weather_data.shape)
print("reshaped weather data:\n",weather_data_reshaped)
```

Summary:-

NumPy allows efficient storage and manipulation of daily temperature readings for a week as an array. This enables quick calculations like the average, maximum, and minimum temperatures across those seven days using vectorized operations.

**3.Sales data analysis.**

**Real-life implementation scenarios:**Sales data analysis.

**import libraries**

```python
import pandas as pd
```

**load the data**

```python
#Load sales data
Sales_data={'Date':['2021-01-01','2021-01-02','2021-01-03','2021-01-04'],
            'store':['Store A','Store B','Store A','Store B'],
            'Sales':[200,150,220,None]}
Sales_data=pd.DataFrame(Sales_data)
Sales_data
```

```python
#example problem for loading data in csv format
df=pd.read_csv("/survey.csv")
df
```

```python
df.isnull().sum()
```

```python
df_new=df
columns_to_fill = ["work_interfere", "self_employed", "state"]
df_new[columns_to_fill] = df_new[columns_to_fill].fillna(method='bfill')
df_new
```

```python
df_new.isnull().sum()
```

step-2:Handling Missing values

```python
#Filling missing sales with the mean Sales Value
Sales_data['Sales']=Sales_data['Sales'].fillna(Sales_data['Sales'].mean())
Sales_data
```

```python
Sales_data.isnull().sum()
```

step-3:-Merge Additional Data

```python
#Additional data:store locations
location_data={'store':['Store A','Store B'],'Location':['New York','Los Angeles']}
location_df=pd.DataFrame(location_data)
location_df
```

```python
#Merging sales data with location data
merged_df=pd.merge(Sales_data,location_df,on='store')
merged_df
```

step-4:-Group and Aggregate Data

```python
#Group by 'Location' and calculate total sales
grouped_sales=merged_df.groupby('store').sum()
grouped_sales
```

```python
grouped_sales=merged_df.groupby('Location').sum()
grouped_sales
```

step-5:- Reshape Data

pivot table:Total sales by Date and location

```python
pivot_table=merged_df.pivot_table(values='Sales',index='Date',columns='Location',aggfunc='sum')
pivot_table
```

```python
#melt the pivot tables back to long format
melted_Sales=pd.melt(pivot_table.reset_index(),id_vars='Date',value_vars=['New York','Los Angeles'],var_name='Location',value_name='Sales')
melted_Sales
```
